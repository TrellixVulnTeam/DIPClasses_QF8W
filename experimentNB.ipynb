```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

pil_im=Image.open('face.jpg')
r,g,b=pil_im.getpixel((1,1))

pil_imgray=pil_im.convert('LA')
plt.imshow(pil_imgray)

plt.savefig("grey.jpg")
```
